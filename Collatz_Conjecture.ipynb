# The Collatz Conjecture

**Introduction :**

The Collatz Conjecture, also known as the 3n + 1 conjecture or the hailstone sequence, is a mathematical problem that involves iterating over a sequence of numbers with a specific set of rules. The conjecture is named after the German mathematician Lothar Collatz, who first proposed it in 1937.

**Explanation :**



1. Start with any positive integer n.
2. If n is even, divide it by 2 $\left(n → \frac{n}{2}\right)$.
3. If n is odd, multiply it by 3 and add 1 $\left(n → 3n + 1\right)$.
4. Repeat the process with the new value of n obtained in the previous step.
5. Continue this iteration until n becomes equal to 1.


The conjecture states that, regardless of the initial positive integer chosen, the sequence of numbers generated will always eventually reach the value 1, at which point it will enter a cycle $\left(6 → 3 → 10 → 5 → 16 → 8 → 4 → 2 → 1\right)$ and continue indefinitely in this cycle.



**Simulation :**

Today, we won't get into the mathematical side of the conjecture (maybe later), but we will look at how numbers' sequences get connected at some point, and try to plot the sequences.

We first import our needed library :

```python
import matplotlib.pyplot as plt
```

We write our simple function to generate the sequence of each number :

```python
def collatz_sequence(n):
    sequence = [n]
    while n != 1:
        if n % 2 == 0:
            n //= 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence
```

We, then, create our function to create all the collatz sequences from 1 to a certain number taken as an parameter :

```python
def generate_collatz_sequences(max_number):
    collatz_sequences = []
    for num in range(1, max_number + 1):
        sequence = collatz_sequence(num)
        collatz_sequences.append(sequence)
    return collatz_sequences
```

We add our function to plot the whole simulation :

```python
def plot_collatz_connections(collatz_sequences, max_number):
    plt.figure(figsize=(16, 16))  # Set a larger figure size
    for num, sequence in enumerate(collatz_sequences, start=1):
        plt.plot(sequence, label=f'Number {num}')

    plt.title(f'Collatz Sequences for Numbers 1 to {max_number}')
    plt.xlabel('Step')
    plt.ylabel('Value')

    # Set x-axis interval to 5
    plt.xticks(range(0, int(max_number*1.4), 4))

    # Calculate the maximum value within all Collatz sequences
    max_value = max(max(sequence) for sequence in collatz_sequences)

    # Set y-axis interval to 100
    plt.yticks(range(0, max_value + 1, 250))

    plt.grid()  # Add a grid
    plt.show()
```

We set our parameters and run our main code :

```python
max_number = 100
collatz_sequences = generate_collatz_sequences(max_number)
plot_collatz_connections(collatz_sequences, max_number)
```
